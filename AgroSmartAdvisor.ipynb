```python
# (2) Load Agmarknet prices (exported by you) & make trend features

import pandas as pd
from pathlib import Path

# Paths
RAW = Path("data/raw")
PROCESSED = Path("data/processed")
RAW.mkdir(parents=True, exist_ok=True)
PROCESSED.mkdir(parents=True, exist_ok=True)

# ✅ Load your downloaded Agmarknet CSV
file_path = r"C:\Users\S\Downloads\9ef84268-d588-465a-a308-a864a43d0070.csv"
prices = pd.read_csv(file_path)

# ✅ First rename columns properly
prices = prices.rename(columns={
    'Arrival_Date': 'date',
    'State': 'state',
    'District': 'district',
    'Market': 'market',
    'Commodity': 'commodity',
    'Variety': 'variety',
    'Grade': 'grade',
    'Min_x0020_Price': 'min_price',
    'Max_x0020_Price': 'max_price',
    'Modal_x0020_Price': 'modal_price'
})

# ✅ Now keep only the relevant columns
cols = ['date','state','district','commodity','min_price','max_price','modal_price']
prices = prices[cols].copy()

# ✅ Convert date column
prices['date'] = pd.to_datetime(prices['date'], errors='coerce', dayfirst=True)

# ✅ Create monthly period
prices['month'] = prices['date'].dt.to_period('M').dt.to_timestamp()

# ✅ Aggregate: monthly average prices per state-district-commodity
monthly = (
    prices.groupby(['state','district','commodity','month'])
    .agg({
        'min_price': 'mean',
        'max_price': 'mean',
        'modal_price': 'mean'
    })
    .reset_index()
)

# ✅ Save cleaned monthly file
out_file = PROCESSED/'prices_monthly.csv'
monthly.to_csv(out_file, index=False)

print(f"✅ Cleaned monthly price trends saved to: {out_file}")
print(monthly.head())
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
import lightgbm as lgb
import numpy as np
import joblib, os

# -------------------------------
# 1) Load Kaggle crop dataset
# -------------------------------
kaggle = pd.read_csv(r"C:\Users\S\Downloads\archive (1)\Crop_recommendation.csv")

# Encode crop labels
crops = sorted(kaggle['label'].unique())
X = kaggle[['N', 'P', 'K', 'ph', 'temperature', 'humidity', 'rainfall']]

# Train one-vs-rest models for crop suitability
suit_models = {}
for crop in crops:
    y = (kaggle['label'] == crop).astype(int)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
    clf = RandomForestClassifier(n_estimators=300, random_state=0)
    clf.fit(X_train, y_train)
    suit_models[crop] = clf

print(f"✅ Trained suitability models for {len(crops)} crops")

# -------------------------------
# 2) Load demand/prices dataset
# -------------------------------
prices = pd.read_csv(r"C:\Users\S\Downloads\9ef84268-d588-465a-a308-a864a43d0070.csv")

# Normalize column names
prices.columns = prices.columns.str.strip().str.lower()
print("🔎 Cleaned price dataset columns:", prices.columns.tolist())

# Rename weird XML-encoded names
prices = prices.rename(columns={
    "min_x0020_price": "min_price",
    "max_x0020_price": "max_price",
    "modal_x0020_price": "modal_price",
    "arrival_date": "arrival_date",
})

# Filter for one state
STATE = "Karnataka"
p = prices[prices['state'].str.contains(STATE, case=False, na=False)].copy()

# Convert arrival_date to datetime
p["arrival_date"] = pd.to_datetime(p["arrival_date"], errors="coerce")

# Create rolling average (3 months)
p = p.sort_values("arrival_date")
p["price_ma3"] = p["modal_price"].rolling(3, min_periods=1).mean()

# Create target: price above median in past 12 months
median_price = p["modal_price"].median()
p["price_above_12m_med"] = (p["modal_price"] > median_price).astype(int)

# Encode commodities
commodities = sorted(p['commodity'].unique())
p['comm_id'] = p['commodity'].map({c: i for i, c in enumerate(commodities)})

# Features + target
Xd = p[['comm_id', 'modal_price', 'price_ma3']].fillna(0)
yd = p['price_above_12m_med']

# Split data
from sklearn.model_selection import train_test_split
Xd_train, Xd_test, yd_train, yd_test = train_test_split(Xd, yd, test_size=0.2, random_state=0)

# Train LightGBM
import lightgbm as lgb
dlgb = lgb.LGBMClassifier(n_estimators=400, learning_rate=0.05, max_depth=-1)
dlgb.fit(Xd_train, yd_train)

print("✅ Demand model trained for", STATE)

# -------------------------------
# 3) Save models
# -------------------------------
os.makedirs("models", exist_ok=True)
joblib.dump(suit_models, "models/suitability_models.pkl")
joblib.dump({"model": dlgb, "commodities": commodities}, "models/demand_model.pkl")

print("💾 Saved suitability_models.pkl and demand_model.pkl")
```

```python
#5) Streamlit app (very small MVP)
import streamlit as st
import pandas as pd
import numpy as np
import joblib

st.set_page_config(page_title='Smart Crops', page_icon='🌾', layout='centered')
st.title('🌾 Smart Crop Recommender (MVP)')

# Inputs
state = st.selectbox('State', ['Karnataka','Maharashtra','Tamil Nadu','Telangana','Andhra Pradesh'])
month = st.selectbox('Month', list(range(1,13)), index=0)

st.subheader('Soil test (or keep defaults)')
N = st.number_input('Nitrogen (N)', 0, 200, 90)
P = st.number_input('Phosphorus (P)', 0, 200, 42)
K = st.number_input('Potassium (K)', 0, 200, 43)
ph = st.slider('pH', 3.5, 9.5, 6.5)

temp = st.slider('Temperature °C', 5.0, 45.0, 25.0)
hum = st.slider('Humidity %', 10, 100, 60)
rain = st.slider('Rainfall (mm)', 0.0, 400.0, 100.0)

if st.button('Recommend crops'):
    # Load models
    suit_models = joblib.load('models/suitability_models.pkl')
    demand_pack = joblib.load('models/demand_model.pkl')
    demand_model = demand_pack['model']
    commodities = demand_pack['commodities']
    comm_map = {c: i for i, c in enumerate(commodities)}

    # Suitability score per crop
    X = np.array([[N, P, K, ph, temp, hum, rain]])
    suit_scores = {crop: mdl.predict_proba(X)[0, 1] for crop, mdl in suit_models.items()}

    # Demand score per crop (state + month)
    state_idx = 0  # 👉 if you only trained Karnataka; later map properly
    demand_X = np.array([[state_idx, month]])
    demand_scores = demand_model.predict_proba(demand_X)[0]

    # Combine scores
    final_scores = {}
    for crop, sscore in suit_scores.items():
        if crop in comm_map:
            dscore = demand_scores[comm_map[crop]]
            final_scores[crop] = 0.5 * sscore + 0.5 * dscore

    # Show top crops
    ranked = sorted(final_scores.items(), key=lambda x: x[1], reverse=True)[:5]
    st.subheader("🌱 Recommended Crops")
    for crop, score in ranked:
        st.write(f"**{crop}** → {score:.2f}")
```
